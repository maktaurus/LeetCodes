**Binary Tree**: are data structures in which each node has at most two children, often referred as left and right children.

Traversal of binary tree:
1)	Depth First Search
a.	Preorder traversal
b.	Inorder traversal
c.	Post order traversal
2)	Breadth First traversal
a.	Level order traversal
