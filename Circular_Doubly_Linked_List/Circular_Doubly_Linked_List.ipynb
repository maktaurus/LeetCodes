# Circular Doubly Linked List
Circular doubly linked list is a more complexed type of data structure in which a node contain pointers to its previous node as well as the next node. Circular doubly linked list doesn't contain NULL in any of the node. The last node of the list contains the address of the first node of the list. The first node of the list also contain address of the last node in its previous pointer.


```python
class Node:
    def __init__(self,val=0,next=None,prev=None):
        self.val = val
        self.next = next
        self.prev = prev

class CircularDoubelyLinkedList:
    def __init_(self):
        self.head = None
        self.tail = None

    def createCDLL(self,value):
        new_node = Node(value)
        self.head = new_node
        self.tail = new_node
        new_node.next = new_node
        new_node.prev = new_node
        return "CDLL is created successfully"

    def __iter__(self):
        temp = self.head
        while temp:
            yield temp
            if temp == self.tail:
                break
            temp = temp.next

    def __len__(self):
        if self.head is None:
            print(f"0 Length")
        else:
            temp = self.head
            length = 0
            while temp.next:
                length += 1
                if temp == self.tail:
                    break
                temp = temp.next
            return length

    def insert(self,index,value):
        if self.head is None:
            return "No CDLL is present"

        new_node = Node(value)
        if index == 0:
            new_node.next = self.head
            self.head = new_node
            self.tail.prev = new_node
            self.tail.next = new_node
            new_node.prev = self.tail
        elif index == self.__len__():
            new_node.prev = self.tail
            new_node.next = self.head
            self.tail.next = new_node
            self.tail = new_node
            self.head.prev = new_node
        elif index == self.__len__()-1:
            prev_node = self.tail.prev
            new_node.next = self.tail
            new_node.prev = prev_node
            prev_node.next = new_node
            self.tail.prev = new_node
        elif index > self.__len__() or index < 0:
            return "Invalid Index Positions"
        else:
            temp = self.head
            for _ in range(index):
                temp = temp.next
            prev_node = temp.prev
            new_node.next = temp
            new_node.prev = prev_node
            temp.prev = new_node
            prev_node.next = new_node

    def traverse(self):
        temp = self.head
        while temp:
            print(temp.val)
            if temp == self.tail:
                break
            temp = temp.next

    def reverse_traverse(self):
        temp = self.tail
        while temp:
            print(temp.val)
            if temp == self.head:
                break
            temp = temp.prev

    def search(self,value):
        if self.head is None:
            return "There is no node in CDLL."
        temp = self.head
        index = 0
        while temp:
            if temp.val == value:
                return f"{temp.val} is a index {index}."
            if temp == self.tail:
                return "The value does not exist in CDLL."
            temp = temp.next
            index += 1

    def remove(self,index):
        if self.head is None:
            return "There is no node in CDLL."
        if index == 0:
            nxt_node = self.head.next
            self.head = nxt_node
            nxt_node.prev = self.tail
            self.tail.next = nxt_node
        elif index == self.__len__()-1:
            prev_node = self.tail.prev
            self.tail = prev_node
            prev_node.next = self.head
            self.head.prev = prev_node
        elif index >= self.__len__() or index < 0:
            return "Index out of range."
        else:
            length = self.__len__()
            pos = length//2
            if index <= pos:
                temp = self.head
                for _ in range(index):
                    temp = temp.next
                prev_node = temp.prev
                nxt_node = temp.next
                prev_node.next = nxt_node
                nxt_node.prev = prev_node
            else:
                temp = self.tail
                for _ in range((self.__len__()-1)-index):
                    temp = temp.prev
                prev_node = temp.prev
                nxt_node = temp.next
                prev_node.next = nxt_node
                nxt_node.prev = prev_node

    def delete_all(self):
        if self.head is None:
            print("There is no node in CDLL")
        else:
            self.tail.next = None
            temp = self.head
            while temp:
                temp.prev = None
                temp = temp.next
            self.head = None
            self.tail = None
```
