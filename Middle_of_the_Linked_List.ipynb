# Doubly Linked List
A doubly linked list or a two-way linked list is a more complex type of linked list that contains a pointer to the next as well as the previous node in sequence.


```python
class Node:
    def __init__(self,value=0,next=None,prev=None):
        self.value = value
        self.next = next
        self.prev = prev

class DobLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0

    def append(self,value):
        new_node = Node(value)
        if self.length == 0:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            new_node.prev = self.tail
            self.tail = new_node
        self.length += 1

    def prepend(self,value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node
        self.length += 1

    def traverse(self):
        curr_note = self.head
        while curr_note:
            print(curr_note.value)
            curr_note = curr_note.next

    def reverse_traverse(self):
        curr_node = self.tail
        while curr_node:
            print(curr_node.value)
            curr_node = curr_node.prev

    def search(self,value):
        temp = self.head
        idx = []
        index = 0
        while temp:
            if temp.value == value:
                idx.append(index)
            temp = temp.next
            index += 1
        if idx:
            return idx
        return -1

    def get(self,index):
        if index < self.length // 2:
            temp = self.head
            for _ in range(index):
                temp = temp.next
        else:
            temp = self.tail
            for _ in range(self.length-1,index,-1):
                temp = temp.prev
        return temp

    def set_val(self, value,index):
        curr_val = self.get(index)
        if curr_val:
            curr_val.value = value
            return True
        return False

    def insert(self,index,value):
        new_node = Node(value)
        if index == 0:
            return self.prepend(value)
        elif index == self.length-1 or index == -1:
            return self.append(value)
        elif index >= self.length or index < -1:
            return False
        else:
            new_node = Node(value)
            curr_node = self.get(index)
            prev_node = self.get(index-1)
            new_node.next = curr_node
            new_node.prev = prev_node
            prev_node.next = new_node
            curr_node.prev = new_node

    def pop_first(self):
        popped_node = self.head
        self.head = self.head.next
        self.head.prev = None
        popped_node.next = None
        self.length -= 1
        return popped_node

    def pop(self):
        popped_node = self.tail
        self.tail = self.tail.prev
        self.tail.next = None
        self.length -= 1
        return popped_node

    def remove(self,index):
        if index == 0:
            return self.pop_first()
        elif index == self.length-1:
            return self.pop()
        elif index >= self.length:
            return None
        else:
            prev_node = self.get(index-1)
            popped_node = self.get(index)
            next_node = self.get(index+1)
            prev_node.next = next_node
            next_node.prev = prev_node
            popped_node.next = popped_node.prev = None
            self.length -= 1
            return popped_node

    def remove_dup(self,value):
        temp = self.head
        dummy = Node()
        dummy.next = temp
        temp.prev = dummy
        prev = dummy

        while temp:
            if temp.value == value and temp == self.head:
                popped_node = temp
                next_node = temp.next
                prev.next = next_node
                next_node.prev = prev
                self.head = next_node
                popped_node.next = popped_node.prev = None
                temp = next_node
                self.length -= 1
            elif temp.value == value and temp == self.tail:
                popped_node = temp
                self.tail = prev
                prev.next = None
                popped_node.prev = None
                self.length -= 1
                break
            elif temp.value == value:
                popped_node = temp
                next_node = temp.next
                prev.next = next_node
                next_node.prev = prev
                temp = next_node
                self.length -= 1
            else:
                prev = temp
                temp = temp.next

    def __str__(self):
        temp = self.head
        ret = ""
        while temp:
            ret += str(temp.value)+" "
            temp = temp.next
        return ret
```
