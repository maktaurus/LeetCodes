# Circular Linked list
A circular linked list is that in which the last node contains the pointer to the first node of the list.


```python
class Node():
    def __init__(self,value=0,next=None):
        self.value = value
        self.next = next

class CSLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0

    def append(self,value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
            new_node.next = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
            new_node.next = self.head
        self.length += 1

    def prepend(self,value):
        new_node = Node(value)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
            new_node.next = new_node
        else:
            new_node.next = self.head
            self.head = new_node
            self.tail.next = new_node
        self.length += 1

    def insert(self,index,value):
        new_node = Node(value)
        temp = self.head
        if index >= self.length and index < -1:
            return None
        elif index == 0:
            self.prepend(value)
        elif index == self.length-1 or index == -1:
            self.append(value)
        else:
            for _ in range(index-1):
                temp = temp.next
            next_node = temp.next
            temp.next = new_node
            new_node.next = next_node
            self.length += 1

    def pop_first(self):
        if self.length == 0:
            return None
        if self.length ==  1:
            self.head = self.tail = None
        else:
            popped_node = self.head
            self.head = self.head.next
            self.tail.next = self.head
            popped_node.next = None
        self.length -= 1
        return popped_node

    def pop(self):
        if self.length == 0:
            return None
        popped_node = self.tail
        if self.length ==  1:
            self.head = self.tail = None
        else:
            temp = self.head
            while temp.next is not self.tail:
                temp = temp.next
            temp.next = self.head
            self.tail = temp
            popped_node.next = None
        self.length -= 1
        return popped_node

    def traverse(self):
        temp = self.head

        while temp:
            print(temp.value)
            if temp == self.tail:
                break
            temp = temp.next

    def get(self,index):
        temp = self.head

        for _ in range(index):
            temp = temp.next
            return temp
        return False

    def set_value(self,index,value):
        curr_val = self.get(index)
        if curr_val:
            curr_val.value = value
            return True
        return False

    def search(self,value):
        temp = self.head
        index = 0

        while temp:
            if temp.value == value:
                return index
            temp = temp.next
            index += 1
            if temp == self.head:
                break
        return False

    def deletefrom(self, index):
        temp = self.head
        if index == 0:
           popped_node = self.head
           self.head = self.head.next
           self.tail.next = self.head
           popped_node.next = None
        elif index == self.length-1 or index == -1:
            popped_node = self.tail
            while temp.next is not self.tail:
                temp = temp.next
            self.tail = temp
            temp.next = self.head
            popped_node.next = None
        else:
            for _ in range(index-1):
                temp = temp.next
            popped_node = temp.next
            temp.next = temp.next.next
            popped_node.next = None

    def remove_dup(self,value):
        temp = self.head
        dummy = Node()
        dummy.next = self.head
        prev = dummy
        index = 0

        while temp:
            if temp.value == value and index == 0:
                next_node = temp.next
                prev.next = next_node
                self.head = next_node
                self.tail.next = next_node
                temp = next_node
                print(index)
            elif temp.value == value and temp == self.tail:
                self.tail = prev
                prev.next = self.head
                break
            elif temp == self.tail:
                break
            elif temp.value == value:
                next_node = temp.next
                prev.next = next_node
                temp = next_node
            else:
                prev = temp
                temp = temp.next
            index += 1

    def delete_all(self):
        self.tail.next = None
        self.head = self.tail = None
        self.length = 0


    def __str__(self):
        temp = self.head
        ret = ""
        while temp:
            ret += str(temp.value)+" "
            if temp == self.tail:
                break
            temp = temp.next
        return ret
```
